```python
import pandas as pd
df=pd.read_csv("horse.csv")
df
```

```python
df.describe()
```

```python
df1=pd.get_dummies(df)
df1=df1.drop_duplicates()
df1
```

```python
x=df1.iloc[:,:].values
x
```

```python
from sklearn.impute import KNNImputer
knn=KNNImputer()
data=knn.fit_transform(x)
data
```

```python
from sklearn.preprocessing import MinMaxScaler
mm=MinMaxScaler()
mm_op=mm.fit_transform(data)
mm_op
```

```python
df2=pd.DataFrame(mm_op)
df2
```

```python
a=df2.iloc[:,0:65].values
a
```

```python
b=df2.iloc[:,67].values
b
```

```python
from sklearn.model_selection import train_test_split
a1,a2,a3,a4=train_test_split(a,b,test_size=0.30,random_state=0)
```

```python
from sklearn.tree import DecisionTreeClassifier
dct=DecisionTreeClassifier(criterion="entropy")
dct.fit(a1,a3)
tree_output=dct.predict(a2)
tree_output
```

```python
from sklearn.metrics import confusion_matrix
log=confusion_matrix(tree_output,a4)
log
```

```python
from sklearn.ensemble import RandomForestClassifier
rc=RandomForestClassifier()
rc.fit(a1,a3)
rf_output=rc.predict(a2)
rf_output
```

```python
from sklearn.metrics import confusion_matrix
rf=confusion_matrix(rf_output,a4)
rf
```
